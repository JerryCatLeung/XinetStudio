## 包 & 库
所有能使用的 $R$ 函数都被包含在一个**库**（Library）中，该库存放在 $R$ 安装文件夹的 $Library$ 目录下。这个目录下包含具有各种功能的**包**（Package），各个包也是按照目录的方式组织起来的。

**包**是函数、数据、预编译代码以一种定义完善的格式组成的集合。

- `install.packages()` 函数可以用来完成包的安装。
- `update.packages()` 函数可以用来更新包。
- `installed.packages()` 函数用来列出以安装的包。
- `search()` 可以用来显示那些包已经加载并可用。
- `library()` 加载以安装的包。

### `apply`

````r
## Compute row and column sums for a matrix:
x <- cbind(x1 = 3, x2 = c(4:1, 2:5))
dimnames(x)[[1]] <- letters[1:8]
apply(x, 2, mean, trim = .2)
col.sums <- apply(x, 2, sum)
row.sums <- apply(x, 1, sum)
rbind(cbind(x, Rtot = row.sums), Ctot = c(col.sums, sum(col.sums)))

stopifnot( apply(x, 2, is.vector))

## Sort the columns of a matrix
apply(x, 2, sort)
````

## 循环结构


?'while'
```r
if(cond) expr
if(cond) cons.expr  else  alt.expr

for(var in seq) expr
while(cond) expr
repeat expr
break
next
```

````r
i = 0
while(i <= 5){
    cat(i, ' ')
    i = i + 1
}
````

````r
i = 0
repeat{
    cat(i, ' ')
    i <- i+1
    if (i>5){
        cat('\n')
        break
    }
}
````

````r
for (i in 1:5){
    cat(i, ' ')
}
````

````r
for(i in 100:999){
    a = floor(i/100)
    b = floor((i-a*100)/10)
    c = i - a * 100 - b * 10
    
    if (a^3 + b^3 + c^3 == i) cat(i, '\n')
}
````

````r
boxplot.stats(mpg)

fivenum(mpg)

x
fivenum(x)

y
fivenum(y)

my.fivenum<-function(x){
    x<-sort(x)
    n <- length(x)
    n4 <- floor((n + 3)/2)/2
    d <- c(1, n4, (n + 1)/2, n + 1 - n4, n)
    return(0.5 * (x[floor(d)] + x[ceiling(d)]))
}

summary(mpg)
````

````r
quantile(mpg)

my.quantile <- function(x) {
    n <- length(x)
    probs = seq(0, 1, 0.25)
    index <- 1 + (n - 1) * probs
    
    lo <- floor(index)
    hi <- ceiling(index)
    
    x <- sort(x, partial = unique(c(lo, hi)))
    qs <- x[lo]
    i <- which(index > lo)
    h <- (index - lo)[i]
    qs[i] <- (1 - h) * qs[i] + h * x[hi[i]]
    return(qs)
}

boxplot(mpg ~ cyl, data = mtcars, main = "Car Mileage Data",
        xlab = "Number of Cylinders", ylab = "Miles/Gallon")
````
