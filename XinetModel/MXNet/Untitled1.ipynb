```python
import numpy as np
import pandas as pd
```

```python
a = np.arange(35).reshape((7, 5))
b = pd.DataFrame(a, columns=5 * ['data'])
b['label'] = [0, 1, 0, 1, None, 2, 0]
b
```

```python
b['data'] 
```

```python
groups = b.groupby('label')
```

```python
groups.mean()
```

```python
groups.apply(lambda x: x.mean())
```

```python
def normalize(df):
    df1 = df[df.columns[:-1]]
    df1 = df1.apply(lambda x: (x - x.mean(0)) / x.var(0))
    df1['label'] = df['label']
    return df1
```

```python
normalize(b)
```

```python
groups.apply(normalize)
```
